```python
import pandas as pd
```

```python
df = pd.read_csv('movies.csv')
df = df.drop(columns=['Unnamed: 0'])
df['overview'] = df['overview'].astype(str)
df['title'] = df['title'].astype(str)
df.head()
```

```python
temp_df = df.copy()
```

```python
df['overview'][0]
```

```python
# Lowercasing
df['overview'] = df['overview'].str.lower()
df['overview'][0]
```

```python
# Remove HTML tags

# df['overview'][0] = '<p> imprisoned in the 1940s for the double murder of his wife and her lover <p>, upstanding banker andy dufresne begins a new life at the shawshank prison, where he puts his accounting skills to work for an amoral warden. during his long stretch in prison, dufresne comes to be admired by the other inmates -- including an older prisoner named red -- for his integrity and unquenchable sense of hope.'
# df['overview'][0]

import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'',text)

df['overview'] = df['overview'].apply(remove_html_tags)
df['overview'][0]
```

```python
# Rmove URLs
def remove_url(text):
  pattern=re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)

df['overview'] = df['overview'].apply(remove_url)
df['overview'][0]
```

```python
# Remove punctuation
import string
exclude = string.punctuation
def remove_punc(text):
  return text.translate(str.maketrans('','',exclude))

df['overview'] = df['overview'].apply(remove_punc)
df['overview'][0]
```

```python
# # Spelling corrction
# from textblob import TextBlob
# def correct_spelling(text):
#   return str(TextBlob(text).correct())

# df['overview'] = df['overview'].apply(correct_spelling)
# df['overview'][0]

# alternative and faster way
!pip install symspellpy
from symspellpy import SymSpell, Verbosity
import pkg_resources

# Setup SymSpell
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)

# Load standard English dictionary
dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt"
)
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)


# Define the cleaning function
def movie_spell_fix(text):
    if not isinstance(text, str) or len(text) < 5:
        return text

    results = sym_spell.lookup_compound(text, max_edit_distance=2, transfer_casing=True)
    return results[0].term

temp_df['overview'] = df['overview'].apply(movie_spell_fix)
temp_df['overview'][0]
```

```python
# Remove stop words
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
  words = text.split()
  filtered_words = [word for word in words if word.lower() not in stop_words]
  return ' '.join(filtered_words)

temp_df['overview'] = df['overview'].apply(remove_stopwords)
temp_df['overview'][0]
```

```python
# Tokenization
import spacy
npl = spacy.load('en_core_web_sm')

def tokenization(text):
  doc = npl(text)
  return [token.text for token in doc]

temp_df['overview'] = df['overview'].apply(tokenization)
temp_df['overview'][0]
```

```python
# Stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem_words(text):
  return ' '.join([ps.stem(word) for word in text.split()])

temp_df['overview'] = df['overview'].apply(stem_words)
temp_df['overview'][0]
```

```python
# Lemmatization

# old way
# from nltk.stem import WordNetLemmatizer
# wordnet_lemmatizer = WordNetLemmatizer()

# def lemmatize_words(text):
#   return ' '.join([wordnet_lemmatizer.lemmatize(word,pos='v') for word in text.split()])

# temp_df['overview'] = df['overview'].apply(lemmatize_words)
# temp_df['overview'][0]

# fastest way
npl = spacy.load('en_core_web_sm',disable=['parser','ner'])
def lemmatize_words(text):
  lemmetize_list = []
  for doc in npl.pipe(text,batch_size=1000,n_process=-1):
    lemmetize_list.append(' '.join([token.lemma_ for token in doc]))
  return lemmetize_list

temp_df['overview'] = lemmatize_words(df['overview'].astype(str).tolist())
print("Done!")
```
